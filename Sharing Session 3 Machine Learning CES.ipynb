## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fs = 1000
f = 5
x = np.arange(fs)
y = np.sin(2 * np.pi * f * x / fs)
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(x, y + np.random.normal(0, 0.1, fs))
```

```python
z = np.linspace(-1, 1, 100) + np.random.normal(0, 0.5, 100)
plt.plot(z)
```
